```python
df['product_id']= df['product_id'].astype('string')
df['product_name']= df['product_name'].astype('string')
df['category']= df['category'].astype('string')

# Clean discounted_price column
df['discounted_price'] = df['discounted_price'].astype(str)
df['discounted_price'] = (
    df['discounted_price']
    .str.replace('₹', '', regex=False)
    .str.replace(',', '', regex=False)
    .str.strip()
)

# Convert safely to numeric
df['discounted_price'] = pd.to_numeric(df['discounted_price'], errors='coerce')

# Clean actual_price column
df['actual_price'] = df['actual_price'].astype(str)
df['actual_price'] = (
    df['actual_price']
    .str.replace('₹', '', regex=False)
    .str.replace(',', '', regex=False)
    .str.strip()
)

# Convert safely to numeric
df['actual_price'] = pd.to_numeric(df['actual_price'], errors='coerce')


df['discount_percentage'] = df['discount_percentage'].astype(str)
df['discount_percentage'] = df['discount_percentage'].str.replace('%', '', regex=False).str.strip()

# Convert safely to numeric
df['discount_percentage'] = pd.to_numeric(df['discount_percentage'], errors='coerce')

df['rating']= pd.to_numeric(df['rating'], errors='coerce')

df['about_product']= df['about_product'].astype('string')
df['user_id']=df['user_id'].astype('string')
df['user_name']= df['user_name'].astype('string')
df['review_id']= df['review_id'].astype('string')
df['review_title']=  df['review_title'].astype('string')
df['review_content']= df['review_content'].astype('string')
df['img_link']= df['img_link'].astype('string')
df['product_link']= df['product_link'].astype('string')
```

Group by 

. Built-in Aggregation Functions

These are the most common and directly supported by pandas:

Function	Description
sum()	Sum of values
mean()	Average of values
median()	Median (middle value)
min()	Minimum value
max()	Maximum value
count()	Number of non-NA values
nunique()	Number of unique values
std()	Standard deviation
var()	Variance
prod()	Product of all values
first()	First value in group
last()	Last value in group
sem()	Standard error of mean
size()	Number of rows in group (including NaN)

```python
df.groupby('category')['actual_price'].sum()
```

```python
df.groupby(['category', 'rating']).agg({'actual_price': ['sum', 'mean'], 'rating_count': 'mean'})
```
